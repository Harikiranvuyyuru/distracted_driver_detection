```python
# encoding: utf-8


import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('driver_imgs_list.csv')
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
print('df shape:', df.shape)
```

```python
df['subject'].value_counts().plot(kind='bar',figsize=(8, 8))  # 每个司机对应的图片数
```

```python
df['classname'].value_counts().plot(kind='bar',figsize=(8, 8))  # 各个类别的图片数
```

```python
plt.figure(figsize=(12, 4))  # width, height

for i in range(12):
    img_path = os.path.join('imgs', 'train', df.iloc[i]['classname'], df.iloc[i]['img'])
    plt.subplot(2, 6, i + 1)  # rows, cols
    img = plt.imread(img_path)
    plt.axis('off')
    plt.imshow(img)
```
